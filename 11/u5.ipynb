```python
# Cargamos datos de gapminder a un df
path = 'https://raw.githubusercontent.com/IDSB-course/m2-sql/master/data/gapminder/gapminder.csv'
gapminder_df = pd.read_csv(path)
```

```python
gapminder_df.tail(12)
```

1. Cómo se distribuye la población en 1952 vs 2007?

```python
df_1952 = gapminder_df[gapminder_df['year'] == 1952]
df_2007 = gapminder_df[gapminder_df['year'] == 2007]
```

```python
sns.histplot( x = np.log(df_2007['pop']))
```

```python
sns.histplot(data = df_2007, x = 'pop')
plt.show()
sns.histplot(data = df_1952, x = 'pop')
plt.show()
```

2. Para argentina, cómo evolucionó el PBI Y la población en el tiempo?

```python
df_argentina = gapminder_df[gapminder_df['country'] == 'Argentina']
```

```python
f, ax = plt.subplots(figsize=(7, 10))
sns.lineplot(data = df_argentina, x = 'year', y = 'gdpPercap')

ax.set_xticks(np.arange(df_argentina.year.min(),df_argentina.year.max(), 5))
```

3. Para el último año disponible, muestre scatterplots con la población y esperanzaa de vida por país coloreados por continente

```python
mask = gapminder_df['year'] == max(gapminder_df['year'])
df_max = gapminder_df[mask]
```

```python
df_max
```

```python
f, ax = plt.subplots(figsize=(7, 7))
ax.set(xscale="log")
sns.scatterplot(x = df_max['pop'], y = df_max['lifeExp'], hue = df_max['continent'], ax=ax)
```

4. Genere boxplots con la esperanza de vida por año




```python
sns.boxplot(data = gapminder_df, y = 'lifeExp', x = 'year')
```

5. Genere violin-plots con la esperanza de vida por año


```python
fig, ax = plt.subplots(figsize=(2.7, 40), layout='constrained')

sns.violinplot(data = gapminder_df, x = 'lifeExp', y = 'country')
```
