## Iteraci√≥n

```python
# iteraci√≥n con while

# sumar todos los valores que ingrese el usuario por teclado.
# el ingreso se finalizar√° cuando el n√∫mero sea negativo

# mostrar la suma total

num_ingresado = int(input("Ingrese n√∫mero a sumar: "))
suma = 0

while num_ingresado >= 0: # mientras sea positivo el num ingresado
  suma = suma + num_ingresado
  num_ingresado = int(input("Ingrese n√∫mero a sumar: "))

print("La suma total es de", suma)
```

```python
# sumar valores de una lista

lista_de_valores = [1, 3, 2, 1]
suma = 0

for valor in lista_de_valores:
  suma += valor # equivale a suma = suma + valor

print("La suma total es de", suma)
```
